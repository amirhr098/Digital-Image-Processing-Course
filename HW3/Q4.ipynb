```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('Noisy_Spine.png',0)
```

```python
average5 = cv2.blur(image,(5,5))
```

حذف نویز های بزرگ

```python
average3 = cv2.blur(average5,(3,3))
```

حذف نویز های کوچکتر

```python
kernel = [[-1,-1,-1],
          [-1,8,-1],
          [-1,-1,-1]]
kernel = np.array(kernel)
laplacian = cv2.filter2D(average3, -1, kernel)
```

```python
laplacian = average3-laplacian
```

برگرداندن نرمی تصویر

```python
plt.figure(figsize=(15,15))
plt.subplot(2,2,1),plt.imshow(image,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(average5,cmap = 'gray')
plt.title('Average 5x5'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(average3,cmap = 'gray')
plt.title('Average 3x3'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
average3 = cv2.blur(image,(3,3))
average5 = cv2.blur(average3,(5,5))
laplacian = cv2.filter2D(average5, -1, kernel)
laplacian = average5-laplacian
plt.figure(figsize=(5,9))
plt.imshow(laplacian, cmap='gray')
plt.xticks([]), plt.yticks([])
plt.show()
```
