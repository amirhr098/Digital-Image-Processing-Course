### 5

```python
blobs = cv2.imread("Blobs.png", 0)
plt.imshow(blobs, cmap='gray')
```

```python
def unit_circle(r):
    def distance(x1, y1, x2, y2):
        return np.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)
    d = 2*r + 1
    mat = np.zeros((d, d))
    rx , ry = d/2, d/2
    for row in range(d):
        for col in range(d):
            dist = distance(rx, ry, row, col)
            if dist <= r:
                mat[row, col] = 1
    return mat.astype("uint8")
```

```python
kernel = unit_circle(28)
closed = cv2.morphologyEx(blobs,cv2.MORPH_CLOSE,kernel)
plt.imshow(closed,cmap="gray")
```

```python
kernel = unit_circle(60)
opened = cv2.morphologyEx(closed,cv2.MORPH_OPEN,kernel)
plt.imshow(opened,cmap="gray")
```

```python
kernel = np.ones((3,3))
gradient = cv2.morphologyEx(opened, cv2.MORPH_GRADIENT, kernel)
plt.imshow(gradient, cmap="gray")
```

```python
plt.imshow(blobs*(255-gradient),cmap="gray")
```
